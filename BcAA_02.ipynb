## Ponto Flutuante ou Decimal (float)

O tipo "float"  utilizado para representar nmeros de ponto flutuante, ou seja, nmeros com
casas decimais. Por exemplo, 3.14, -2.71, 0.5 so todos exemplos de nmeros em ponto flutuante.
Os nmeros de ponto flutuante so representados seguindo o padro IEEE 754, que define a
forma como os computadores armazenam e realizam operaes com nmeros em ponto
flutuante.
Em Python, os nmeros de ponto flutuante so criados automaticamente quando utiliza-se casas
decimais em um nmero, ou pode ser criado explicitamente utilizando o formato "float(valor).

- Ponto Flutuante ou Decimal (float)

Criando nmeros de ponto flutuante

Alguns exemplos de como criar nmeros de ponto flutuante em Python:

````python
# Exemplos de nmeros de ponto flutuante
numero1 = 3.14
numero2 = -2.71
numero3 = 0.5
````

- Operaes com nmeros de ponto flutuante

Podemos realizar todas as operaes matemticas bsicas com nmeros de ponto flutuante.

````python
# Exemplos de operaes com nmeros de ponto flutuante
soma = numero1 + numero2
subtracao = numero1 - numero3
multiplicacao = numero2 * numero3
divisao = numero1 / numero3
potencia = numero1 ** 2
````

exatido.

Por exemplo:

````python
resultado = 0.1 + 0.2
print(resultado) # O resultado ser 0.30000000000000004

# Para contornar problemas de preciso,  comum utilizar funes de arredondamento, como
# round(), para obter um resultado mais adequado s necessidades do clculo.
````

Em alguns casos, pode ser necessrio converter um valor "float" para outro tipo de dado, como
"int" ou "str". Para fazer isso, podemos usar as funes int() e str().

````python
# Converso para int
numero_inteiro = int(numero1)
# Converso para str
numero_str = str(numero2)
````

- Funes matemticas embutidas:

Python oferece diversas funes matemticas embutidas que podem ser usadas com nmeros

````python
numeros = [1.5, 2.3, 0.8, -2.0]
valor_absoluto = abs(numero2)
maior_numero = max(numeros)
soma_total = sum(numeros)
````

## Complexo (complex)

O tipo complex  utilizado para representar nmeros complexos, que so compostos por uma
parte real e uma parte imaginria. Um nmero complexo  escrito na forma "a + bj", onde "a"  a
parte real, "b"  a parte imaginria e "j"  a unidade imaginria ((-1)).
Os nmeros complexos so utilizados para lidar com situaes matemticas que envolvem
razes negativas e problemas em vrias reas, como engenharia, fsica e cincia da computao.
Em Python, os nmeros complexos so criados utilizando o formato "complex(a, b)".
- Criando nmeros complexos

Vamos ver alguns exemplos de como criar nmeros complexos em Python:

````python
# Exemplos de nmeros complexos
complexo1 = complex(3, 4) # Representa o nmero 3 + 4j
complexo2 = complex(-2, 1) # Representa o nmero -2 + 1j
complexo3 = 1 + 2j # Podemos tambm criar usando a notao direta
````

- Nmeros complexos - Operaes

Podemos realizar vrias operaes matemticas com nmeros complexos, incluindo adio,
subtrao, multiplicao, diviso e potenciao.

````python
# Exemplos de operaes com nmeros complexos
soma = complexo1 + complexo2
subtracao = complexo1 - complexo3
multiplicacao = complexo1 * complexo2
divisao = complexo2 / complexo3
potencia = complexo3 ** 2
````

- Nmeros complexos - Acessando a parte real e imaginria

Para acessar a parte real e imaginria de um nmero complexo, podemos usar as propriedades
"real" e "imag".

````python
# Exemplos de acesso  parte real e imaginria
parte_real = complexo1.real
parte_imaginaria = complexo1.imag
````

- Nmeros complexos - Funes matemticas e complexas

Python oferece algumas funes matemticas e complexas especficas para trabalhar com
nmeros complexos, como abs(), conjugate(), phase().

````python
# Exemplos de funes matemticas e complexas
modulo = abs(complexo2)
conjugado = complexo1.conjugate()
fase = phase(complexo3) # Retorna o ngulo em radianos
````

- Nmeros complexos - Converso entre tipos de dados

Em alguns casos, pode ser necessrio converter um valor "complex" para outro tipo de dado,
como "int", "float" ou "str". Para fazer isso, podemos usar as funes "int()", "float()" e "str()".

````python
# Converso para int e float
parte_real_inteira = int(complexo1.real)
parte_imaginaria_float = float(complexo1.imag)
# Converso para str
complexo_str = str(complexo2)
````

## Listas (list)

O tipo "list"  utilizado para representar listas, que so colees ordenadas e mutveis de
elementos. As listas so extremamente versteis e amplamente utilizadas em Python para
armazenar conjuntos de dados relacionados.

As listas so conjuntos de elementos que podem ser de diferentes tipos de dados, como
nmeros, strings ou at outras listas. Em Python, as listas so criadas utilizando colchetes [] e os
elementos so separados por vrgulas.

````python
# Exemplos de listas
lista_numeros = [1, 2, 3, 4, 5]
lista_frutas = ['ma', 'banana',
'laranja']
lista_mista = [10, 'python', True]
````

Podemos acessar elementos individuais de uma lista utilizando ndices. Os ndices em Python

````python
# Exemplo de acesso a elementos em uma lista
numeros = [10, 20, 30, 40, 50]
primeiro_elemento = numeros[0] # 10
segundo_elemento = numeros[1] # 20
ultimo_elemento = numeros[-1] # 50 (ltimo elemento da lista)
````

- Listas - Operaes

````python
# Exemplos de operaes com listas
lista_a = [1, 2, 3]
lista_b = [4, 5, 6]

# Adio de elementos
lista_a.append(4) # [1, 2, 3, 4]
lista_b.insert(0, 0) # [0, 4, 5, 6]

# Remoo de elementos
lista_a.remove(2) # [1, 3]
elemento_removido = lista_b.pop(2) # elemento_removido = 5, lista_b = [0,4, 6]

# Concatenao de listas
lista_concatenada = lista_a + lista_b # [1, 3, 0, 4, 6]

# Fatiamento (slicing)
sublista = lista_concatenada[1:4] # [3, 0, 4]
````

## Tuplas (tuple)

O tipo "tupla"  utilizado para representar colees ordenadas e imutveis de elementos. As
tuplas so semelhantes s listas, mas a principal diferena  que as tuplas no podem ser
modificadas aps a sua criao, tornando uma escolha adequada para dados que no devem
ser alterados.

As tuplas so semelhantes s listas, mas so criadas utilizando parnteses () em vez de
colchetes []. A principal caracterstica das tuplas  que elas so imutveis, o que significa que os
elementos no podem ser adicionados, removidos ou alterados aps a criao.

- Tuplas- Criao

````python
# Exemplos de tuplas
tupla_numeros = (1, 2, 3, 4, 5)
tupla_frutas = ('ma', 'banana', 'laranja')
tupla_mista = (10, 'python', True)
````

- Tuplas - Acessando elementos

Assim como nas listas, podemos acessar elementos individuais de uma tupla utilizando ndices.

````python
# Exemplo de acesso a elementos em uma tupla
numeros = (10, 20, 30, 40, 50)
primeiro_elemento = numeros[0] # 10
segundo_elemento = numeros[1] # 20
ultimo_elemento = numeros[-1] # 50 (ltimo elemento da tupla)
````

- Tuplas - Operaes

Como as tuplas so imutveis, algumas operaes que so vlidas para listas no so
permitidas para tuplas. No entanto, podemos realizar operaes como concatenao e
repetio.

````python
# Exemplos de operaes com tuplas
tupla_a = (1, 2, 3)
tupla_b = (4, 5, 6)

# Concatenao de tuplas
tupla_concatenada = tupla_a + tupla_b # (1, 2, 3, 4, 5, 6)

# Repetio de tupla
tupla_repetida = tupla_a * 3 # (1, 2, 3, 1, 2, 3, 1, 2, 3)
````

## Dicionrios (dict)

O tipo "dicionrio"  utilizado para representar colees de elementos que so armazenados em
pares chave-valor. Os dicionrios so estruturas de dados muito poderosas e versteis,
permitindo a organizao e recuperao eficiente de informaes.

Os dicionrios em Python so colees no ordenadas de elementos. Cada elemento 
representado por um par chave-valor, onde a "chave"  um identificador nico e o "valor"  a
informao associada quela chave. Os dicionrios so criados utilizando chaves {} e os pares
chave-valor so separados por dois pontos (:).

- Dicionrios - Criao

````python
# Exemplos de dicionrios
dicionario_pessoas = {
'nome': 'Joo',
'idade': 30,
'cidade': 'So Paulo'
}
dicionario_notas = {
'Matemtica': 9.5,
'Cincias': 8.0,
'Histria': 7.5
}
````

- Dicionrios - Acessando elementos

Podemos acessar os valores associados s chaves de um dicionrio utilizando as chaves como
ndices.

````python
# Exemplo de acesso a elementos em um dicionrio
pessoa = {
'nome': 'Maria',
'idade': 25,
'cidade': 'Rio de Janeiro'
}

nome = pessoa['nome'] # 'Maria'
idade = pessoa['idade'] # 25
cidade = pessoa['cidade'] # 'Rio de Janeiro'
````

````python
# Exemplo de operaes com dicionrios
pessoa = {
'nome': 'Joo',
'idade': 30,
'cidade': 'So Paulo'
}

#Adicionar novo elemento
pessoa['profissao'] = 'Engenheiro'

# Modificar valor associado a uma chave
pessoa['idade'] = 31

# Remover elemento
del pessoa['cidade']
````

- Dicionrios - Funes teis

keys(), values(), items().

````python
# Exemplos de funes teis para dicionrios
pessoa = {
'nome': 'Joo',
'idade': 30,
'cidade': 'So Paulo'
}

tamanho = len(pessoa) # Retorna o nmero de pares chave-valor (3)
chaves = pessoa.keys() # Retorna as chaves do dicionrio ('nome', 'idade', 'cidade')
valores = pessoa.values() # Retorna os valores do dicionrio ('Joo', 30, 'So Paulo')
pares = pessoa.items() # Retorna uma lista de tuplas com os pares chave-valor
````

````python
# adio
a = 5
b = 3
resultado = a + b # Resultado: 8

# subtrao
a = 10
b = 7
resultado = a - b # Resultado: 3

# multiplicao
a = 3
b = 4
resultado = a * b # Resultado: 12

# diviso
a = 10
b = 3
resultado = a / b # Resultado: 3.3333333333333335

# diviso inteira
a = 10
b = 3
resultado = a // b # Resultado: 3

# resto da diviso
a = 10
b = 3
resto = a % b # Resultado: 1

# potncia
a = 2
b = 3
resultado = a ** b # Resultado: 8

# precedncia dos operadores
resultado = 2 + 3 * 4 # Resultado: 14 (multiplicao antes da adio)
resultado = (2 + 3) * 4 # Resultado: 20 (parnteses para alterar a ordem de avaliao)
````

````python
# Exemplo de loop "for" com lista de tuplas
pessoas = [('Joo', 25), ('Maria', 30), ('Pedro', 22)]

for nome, idade in pessoas:
  print(f"{nome} tem {idade} anos.")
````

- Loop for - Dicionrio

````python
# Exemplo de loop "for" com dicionrio
idades = {'Joo': 25, 'Maria': 30, 'Pedro': 22}

# Iterao sobre as chaves
for nome in idades:
  print(nome)

# Iterao sobre os valores
for idade in idades.values():
  print(idade)
````

- Loop for - Dicionrio e Enumerando elementos com "enumerate()"

````python
# Exemplo de loop "for" com dicionrio
idades = {'Joo': 25, 'Maria': 30, 'Pedro': 22}

# Iterao sobre as chaves
for nome in idades:
  print(nome)

# Iterao sobre os valores
for idade in idades.values():
  print(idade)
````

````python
# Exemplo de uso de "enumerate()"
frutas = ['ma', 'banana', 'laranja']

for indice, fruta in enumerate(frutas):
  print(f"ndice: {indice}, Fruta: {fruta}")
````

## Estruturas de repetio Loop while

- Loop while - Sintaxe

```
while condio:
  # Cdigo a ser executado enquanto a condio for verdadeira
  ```

````python
# while condio:
#   # Cdigo a ser executado enquanto a condio for verdadeira
````

- Loop while - Execuo

O loop "while" continuar a ser executado enquanto a condio especificada for avaliada como
True.  importante garantir que a condio seja eventualmente False, caso contrrio, o loop
continuar indefinidamente, resultando em um loop infinito.

````python
# Exemplo de loop "while"
contador = 0

while contador < 5:
  print(contador)
  contador += 1
````

- Loop while - Loop infinito e o comando "break"

````python
# Exemplo de loop "while" com "break"
contador = 0
while True:
  print(contador)
  contador += 1
  if contador >= 5:
    break
````

- Loop while - Comando "continue"

````python
contador = 0
while contador < 5:
  contador += 1
  if contador == 3:
    continue # Pula a iterao quando contador == 3
  print(contador)
````

- Loop while - Iterando sobre listas

````python
# Exemplo de loop "while" para iterar sobre lista
frutas = ['ma', 'banana', 'laranja']

indice = 0

while indice < len(frutas):
  print(frutas[indice])
  indice += 1
````

## Conjuntos

Para criar um conjunto em Python, utilizamos chaves {} ou a funo set().

Vamos ver alguns exemplos:

````python
# Exemplo 1: Conjunto com chaves {}
frutas = {'ma', 'banana', 'laranja'}

# Exemplo 2: Conjunto com a funo set()
numeros = set([1, 2, 3, 4, 5])
````

- Conjuntos Operaes

````python
conj1 = {1, 2, 3}
conj2 = {3, 4, 5}

# Adicionar elemento
conj1.add(4) # conj1 agora  {1, 2, 3, 4}

# Remover elemento
conj1.remove(2) # conj1 agora  {1, 3, 4}

# Verificar a existncia de um elemento
if 3 in conj1:
  print("O nmero 3 est no conjunto conj1.")

# Unio de conjuntos
uniao = conj1.union(conj2) # uniao  {1, 3, 4, 5}

# Interseo de conjuntos
intersecao = conj1.intersection(conj2) # intersecao  {3}

# Diferena entre conjuntos
diferenca = conj1.difference(conj2) # diferenca  {1, 4}
````

- Tcnicas de iterao

Ao iterar sobre dicionrios, a chave e o valor correspondente podem ser obtidos utilizando o
mtodo items().

````python
knights = {'gallahad': 'the pure', 'robin': 'the brave'}

for k, v in knights.items():
  print(k, v)

# Resultado
# gallahad the pure
# robin the brave
````
